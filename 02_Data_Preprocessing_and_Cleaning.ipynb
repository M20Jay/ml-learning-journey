## Section 1 — Import Libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print("library successfully loaded")
````

## Section 2 — Loading Data

````python
df=pd.read_csv("titanic.csv")
print(df.shape)
print("\n First 5 rows")
print(df.head())
````

## Section 3 — Inspecting the Data

````python
# Data types and non-null counts
print("=== Dataset Info ===")
df.info()
````

````python
# Summary statistics
print("=== Summary Statistics ===")
df.describe()
````

````python
# Missing value per column
print("=== Missing Values ===")
missing = df.isnull().sum()
missing_percentage = df.isnull().sum()/len(df)*100
missing_table= pd.DataFrame({
    "Missing Values": missing,
    "Percentage": missing_percentage
})

print(missing_table[missing_table["Missing Values"]>0])
````

## Section 4 — Handling Missing Values

````python
# Fill Age with median — right skewed distribution
df['Age'].fillna(df['Age'].median(),inplace=True)
````

````python
# Drop Cabin — 77% missing, too many to fill
df.drop(columns=["Cabin"], inplace = True)
````

````python
# Fill Embarked with mode — only 2 missing
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print(" Embarked missing values:", df["Embarked"].isnull().sum())
````

````python
# Confirm no missing values remain
print("=== Missing Values After Cleaning ===")
print(df.isnull().sum())
````

## Section 5 — Encoding Categorical Variables

````python
# Encode Sex — female=0, male=1
df['Sex']= df['Sex'].map({'female':1, 'male':0})
print("Sex encoded:")
print(df["Sex"].value_counts())
````

````python
# One Hot Encode Embarked
df= pd.get_dummies(df, columns=["Embarked"], drop_first =True)
print(" Embarked Encoded:")
print(df.columns.tolist())
````

````python
# Confirm encoding
print("=== Dataset After Encoding ===")
print(df.shape)
print(df.head())
````

````python
# Checking current Sex values
print("Sex unique values:")
print(df['Sex'].unique())
````

````python
#Checking current Embarked Values
#print(df['Embarked'].unique())
````

````python
#Confirm everything:
print("=== Dataset After Encoding ===")
print("Shape:", df.shape)
print("\nMissing values:")
print(df.isnull().sum())
print("\nFirst 3 rows:")
df.head(3)
````

````python
# Force drop Cabin
if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)
    print("Cabin dropped ✅")
else:
    print("Cabin already gone ✅")

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
````

## Section 6 — Feature Scaling

````python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
df[['Age', 'Fare']]= scaler.fit_transform(df[['Age','Fare']])
print("===After Scaling")
print(df[["Age", "Fare"]].describe().round(2))
````

## Section 7 — Saving the Clean Dataset

````python
# Save cleaned Dataset
df.to_csv('titanic_clean.csv', index= False)

print("===Dataset saved===")
print("File: titanic_clean.csv ✅")
print("Shape :", df.shape)
print("Colomns:", df.columns.tolist())
````
