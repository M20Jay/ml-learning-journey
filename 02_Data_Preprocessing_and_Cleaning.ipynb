## Section 1 — Import Libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print("library successfully loaded")
````

## Section 2 — Loading Data

````python
df = pd.read_csv('titanic.csv')
print(df.shape)
print("\n First 5 rows")
print(df.head())
````

## Section 3 — Inspecting the Data

````python
# Data types and non-null counts
print("=== Dataset Info ===")
df.info()
````

````python
# Summary statistics
print("=== Summary Statistics ===")
df.describe()
````

````python
# Missing value per column
print("=== Missing Values ===")
missing = df.isnull().sum()
missing_percentage = df.isnull().sum()/len(df)*100
missing_table= pd.DataFrame({
    "Missing Values": missing,
    "Percentage": missing_percentage
})

print(missing_table[missing_table["Missing Values"]>0])
````

## Section 4 — Handling Missing Values

````python
# Fill Age with median — right skewed distribution
df['Age'].fillna(df['Age'].median(),inplace=True)
````

````python
# Drop Cabin — 77% missing, too many to fill
# df.dropna(columns=["Cabin"], inplace = True)
````

````python
# Fill Embarked with mode — only 2 missing
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print(" Embarked missing values:", df["Embarked"].isnull().sum())
````

````python
# Confirm no missing values remain
print("=== Missing Values After Cleaning ===")
print(df.isnull().sum())
````

## Section 5 — Encoding Categorical Variables

````python
# Encode Sex — female=0, male=1
df['Sex']= df['Sex'].map({'famale':1, 'male':0})
print("Sex encoded:")
print(df["Sex"].value_counts())
````

````python
# One Hot Encode Embarked
#df= pd.get_dummies(df, columns=["Embarked"], drop_first =True)
print(" Embarked Encoded:")
print(df.columns.tolist())
````
